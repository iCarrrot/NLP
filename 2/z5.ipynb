```python
import re
import numpy as np
import os
os.sys.path.append('../1/')
from z2 import loader
from math import log
import sys
import heapq

word2tag = dict()
tag2word = dict()
def stringNorm(sent, num=False):
    regex = re.compile(f'[,\.!?:;\'{"0-9" if not num else ""}\*\-“…\(\)„”—»«–––=\[\]’]')
    return regex.sub('',sent.lower())

def bigrams2unigrams(bigrams):
    return {w1: sum([float(bigrams[w1][w2]) for w2 in bigrams[w1]])/2 for w1 in bigrams}

with open("data/supertags.txt") as tags:
    for line in tags:
        word, tag = stringNorm(line, num=True).split()
        word2tag[word] = tag
        if tag in tag2word:
            tag2word[tag].append(word)
        else:
            tag2word[tag] = [word]
            

bigrams = loader('../1/poleval_2grams.txt', cut = 10)
unigrams = bigrams2unigrams(bigrams)
```

```python
uniSum = sum([float(unigrams[x]) for x in unigrams])

PMI = lambda w1, w2: log(
    float(bigrams[w1][w2]) * uniSum / (unigrams[w1] * unigrams[w1]) 
    + sys.float_info.min)

sameTag = lambda w1, w2: (5 * int(word2tag[w1] == word2tag[w2]) + 1 
            if w1 in word2tag and w2 in word2tag 
            else 0)
        
wordValue = lambda word, x: PMI(word, x) + 15 * sameTag(word,x) + 5 * len(x)
```

```python
def findSimilarWords(word, k):
    return [
        f"{x}:       \t\t{wordValue(word,x)}"
        for x in heapq.nlargest(
            k, 
            bigrams[word], 
            key=lambda x: wordValue(word, x)
        )
    ]
```

```python
slowa = ['dom', 'koń', 'telefon', 'pakiet', 'prowadził', 'komputer', 'w', 'niebieski', 'złoty', 'zdobyć']
nln = '\n'
for i in slowa:
    print(f'{i}:\n{nln.join(findSimilarWords(i, 10))}, \n\n')
```
