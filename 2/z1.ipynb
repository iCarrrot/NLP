```python
from collections import Counter
import re
import operator
from math import log as ln

def readData(filename):
    with open(filename) as f:
        file = f.read()
        sents = list(filter(lambda x: len(x)>2, re.split('\?|\.|\!|\n|…',file)))

        sentLen = len(list(filter(lambda x: 20<len(x.split()), sents)))
        
#         print(len(list(filter(lambda x: x>20, lens)))/len(sents))
        regex = re.compile('[,\.!?:;\'0-9\*\-“…\(\)„”—»«–––=\[\]’]')
        fileSplit = regex.sub('',file.lower()).split()
    return fileSplit, sentLen / len(sents), (sentLen, len(sents)-sentLen)

def create_data(filename):
    fileSplit, ppb, _ = readData(filename)
    wordcount = Counter(fileSplit)
    for i in wordcount:
        if not i.isalpha() and i != ' ': 
            print(i)
    return dict(wordcount), len(fileSplit), ppb

def NaiveBayes(test, classInfo, sents):
    posProb = {i:ln(classInfo[i][1]) 
               + ln(classInfo[i][2] ) * sents[0]
               + ln((1 - classInfo[i][2])) * sents[1] for i in 'POS'}
    for w in test:
        for i in 'POS':
            if w in classInfo[i][0]:
#                 print(i, w, classInfo[i][0][w]/float(classInfo[i][1]))
                posProb[i] += ln(classInfo[i][0][w]/float(classInfo[i][1]))
            else:
#                 print(i, w, 'nope', 1./classInfo[i][1]/100. )
                posProb[i] += ln(1./classInfo[i][1]/100.)
    
    return max(posProb.items(), key=operator.itemgetter(1))[0], posProb
```

```python
dP, cP, ppbP = create_data("dane_pozytywistyczne/korpus_prusa.txt")
dO, cO, ppbO = create_data("dane_pozytywistyczne/korpus_orzeszkowej.txt")
dS, cS, ppbS = create_data("dane_pozytywistyczne/korpus_sienkiewicza.txt")

classInfo = {'P':(dP, cP, ppbP),'O':(dO, cO, ppbO),'S':(dS, cS, ppbS),}
for i in ['']+list(range(1,23,2)):
    test, _, sents = readData(f"dane_pozytywistyczne/testy1/test_orzeszkowej{i}.txt")
    result, rest = NaiveBayes(test, classInfo, sents)
    print(result)
    if result!='O':
        print(rest)
print('\n\n\n')
for i in list(range(1,55,2)):
    test, _, sents = readData(f"dane_pozytywistyczne/testy1/test_sienkiewicza{i}.txt")
    result, rest = NaiveBayes(test, classInfo, sents)
    print(result)
    if result!='S':
        print(rest)
        
print('\n\n\n')
for i in list(range(0,42,2)):
    test, _, sents = readData(f"dane_pozytywistyczne/testy1/test_prusa{i}.txt")
    result, rest = NaiveBayes(test, classInfo, sents)
    print(result)
    if result!='P':
        print(rest)
```
