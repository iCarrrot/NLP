```python
SUF = 4
sentences = [
    'judyta dała wczoraj Stefanowi czekoladki',
    'babuleńka miała dwa rogate koziołki',
    'wczoraj wieczorem spotkałem pewną piękną kobietę',
    'Idę kupić pięć litrów mleka',
    'Zawsze lubiłem kupować lody'
]
sentencesSets = [set(x.lower().split(' ')) for x in sentences]
sufix = lambda x: x[-SUF:]
# sufix(sentences[1].split()[0])
```

```python
def loader(fname, top = float('inf'),sX = False):
#     print(sentencesSets)
    dataS = []
    data = [{key:{key2:0 for key2 in sen if key2!=key}  for key in sen} for sen in sentencesSets ]
    if sX:
        dataS = [{sufix(key):{sufix(key2):0 for key2 in sen}  for key in sen} for sen in sentencesSets ]
    
#         for i in range(len(dataS)):
#             print(i, dataS[i])
 
    with open(fname, 'r') as f:
        it=0
        for x in f:
            if not it%1000000:
                print(it)
            it+=1
            if it > top:
                break
            num, key, key2 = x.lower().split()
            for i in range(len(sentencesSets)):
                if key in sentencesSets[i] and  key2 in sentencesSets[i] and key2!=key:
                    data[i][key][key2] = int(num)
                if sX and sufix(key) in dataS[i] and  sufix(key2) in dataS[i] and sufix(key2)!=sufix(key):
 
                    dataS[i][sufix(key)][sufix(key2)] += int(num)
                
    return data, dataS
```

```python
sen2gram, sen2gramS = loader('poleval_2grams.txt',sX=True)
```

```python
# print(sen2gram, '\n\n\n')
import itertools
for i in range(len(sentencesSets)):
    perms = list(itertools.permutations(list(sentencesSets[i])))
    res = []
    for perm in perms:
        s = 0
        for k in range(len(perm)-1):
            s+=int(sen2gram[i][perm[k]][perm[k+1]])
        res.append((s,' '.join(perm)),)
    res.sort()
    for i in res[-15:]:
        print(i[0], i[1])
```
