```python
def weightList(dic):
#     print(dic)
    res = []
    for k in list(dic):
        res+=int(dic[k])*[k]
#         print(res)
    return res
```

```python
sentence = []
newKey = random.choice(weightList(pol2gram['<BOS>']))
sentence.append(newKey)
# print(newKey)
while newKey in pol2gram:
    newKey = random.choice(weightList(pol2gram[newKey]))
#     print(newKey)
    sentence.append(newKey)
print(' '.join(sentence))
```
